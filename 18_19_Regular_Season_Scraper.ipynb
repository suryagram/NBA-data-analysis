```python
def get_nba_schedule(api_token, token_type, season_start_yr, output_file, api_version='v7',
                     language='en', season_type='REG'):
    api_url = f"http://api.sportradar.us/nba/{token_type}/{api_version}/{language}/games/{season_start_yr}/{season_type}/schedule.json?api_key={api_token}"
    schedule_headers = {
    'Content-Type': 'application/json;charset=utf-8',
    'Content-Language': 'en',
    'Accept-Ranges': 'bytes',
    'Age': '0',
    'Cache-Control': 'public, must-revalidate, max-age=11'}
    schedule_response = requests.get(api_url, headers=schedule_headers)
    if schedule_response.status_code == 200:
        schedule_data = json.loads(schedule_response.content.decode('utf-8'))
    with open(output_file, 'w') as outfile:
        json.dump(schedule_data, outfile, indent=4)
```
